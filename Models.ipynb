<h3>Model Training</h3>

```python
X = df.drop(['final_grade', 'G3','school','traveltime','higher','age','freetime','address','Pstatus','Dalc','Walc','famrel','romantic','nursery','schoolsup','famsize','reason','Mjob','Fedu','Fjob'], axis = 1)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, df['final_grade'], test_size=0.3,random_state=100) # 70% training and 30% test
```

```python
X.columns
```

```python
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC() 
```

```python
y_train
```

```python
#Train the model using the training sets
clf.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_test, y_pred)*100)
```
